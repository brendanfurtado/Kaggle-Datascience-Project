```python
import pandas as pd

# This file takes the output of kaggle accuracy and outputs 
data1 = pd.read_csv("prediction_voted.csv")
data2 = pd.read_csv("prediction_voted2.csv")
data3 = pd.read_csv("prediction_voted3.csv")
data4 = pd.read_csv("prediction_voted4.csv")


y_pred1 = data1[data1.columns[-1]]
y_pred2 = data2[data2.columns[-1]]
y_pred3 = data3[data3.columns[-1]]
y_pred4 = data4[data4.columns[-1]]
```

```python
y_pred = []
for i in range(len(y_pred1)):
    if y_pred1[i] == y_pred2[i] == y_pred3[i] == y_pred4[i]:
        y_pred.append(y_pred4[i])
        
    elif y_pred1[i] == y_pred2[i] == y_pred3[i] != y_pred4[i]:
        y_pred.append(y_pred1[i])
        
    elif y_pred1[i] == y_pred3[i] == y_pred4[i] != y_pred2[i]:
        y_pred.append(y_pred1[i])
        
    elif y_pred1[i] == y_pred2[i] == y_pred4[i] != y_pred3[i]:
        y_pred.append(y_pred2[i])
        
    elif y_pred1[i] != y_pred2[i] == y_pred3[i] == y_pred4[i]:
        y_pred.append(y_pred2[i])
        
    elif y_pred1[i] == y_pred2[i] != y_pred3[i] == y_pred4[i]:
        y_pred.append(y_pred1[i])
        
    elif y_pred1[i] == y_pred3[i] != y_pred2[i] == y_pred4[i]:
        y_pred.append(y_pred1[i])   
        
    elif y_pred1[i] == y_pred4[i] != y_pred2[i] == y_pred3[i]:
        y_pred.append(y_pred1[i])
        
    elif y_pred1[i] == y_pred2[i] != y_pred3[i] != y_pred4[i]:
        y_pred.append(y_pred1[i])
        
    elif y_pred1[i] == y_pred3[i] != y_pred4[i] != y_pred2[i]:
        y_pred.append(y_pred1[i])
        
    elif y_pred1[i] == y_pred4[i] != y_pred2[i] != y_pred3[i]:
        y_pred.append(y_pred1[i])
        
    elif y_pred2[i] == y_pred3[i] != y_pred1[i] != y_pred4[i]:
        y_pred.append(y_pred2[i])    
        
    elif y_pred2[i] == y_pred4[i] != y_pred1[i] != y_pred3[i]:
        y_pred.append(y_pred2[i]) 
        
    elif y_pred3[i] == y_pred4[i] != y_pred1[i] != y_pred2[i]:
        y_pred.append(y_pred3[i])
    
    elif y_pred3[i] != y_pred4[i] != y_pred1[i] != y_pred2[i]:
        y_pred.append(y_pred1[i])
```

```python
import numpy as np
Y_merged = np.array(y_pred)

np.savetxt("merged_prediction2.csv", np.dstack((np.arange(0, Y_merged.size),Y_merged))[0],"%s,%s",header="Id,Category")
```
